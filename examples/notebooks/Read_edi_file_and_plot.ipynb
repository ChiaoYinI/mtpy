## plot mt response

```python
#Supress default INFO logging
import logging
logger = logging.getLogger()
logger.setLevel(logging.CRITICAL)
logger.propagate = False

mt_plot = mt_obj.plot_mt_response()
```

*plot phase tensors*

```python
mt_plot.plot_pt = 'y'
mt_plot.redraw_plot()
```

*plot strike angle*

```python
mt_plot.plot_strike = 'ytip'
mt_plot.redraw_plot()
```

```python
mt_plot.plot_strike = 'ytip'
mt_plot.plot_skew = 'y'
mt_plot.plot_tipper='yri'  #?? No Tipper data for station pb40

mt_plot.redraw_plot()
```
