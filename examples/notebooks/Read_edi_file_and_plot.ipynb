# Read EDI file and plot response

## import modules

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['savefig.dpi'] = 300
```

```python
import mtpy.core.mt as mt
```

```python
import os
DATA_DIR=r'E:\GitHubz\mtpy2\examples\data'
edi_fn = os.path.join(DATA_DIR, 'edi_files\pb40c.edi')
mt_obj = mt.MT(edi_fn)
```

*If you need help*

```python
help(mt_obj)
```

## plot mt response

```python
mt_plot = mt_obj.plot_mt_response()
```

*plot phase tensors*

```python
mt_plot.plot_pt = 'y'
mt_plot.redraw_plot()
```

*plot strike angle*

```python
mt_plot.plot_strike = 'ytip'
mt_plot.redraw_plot()
```

*plot all 4 components of the impedance tensor*

```python
mt_plot.plot_num = 2
mt_plot.redraw_plot()
```
