## import modules

```python
import os

%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['savefig.dpi'] = 300
```

```python
import mtpy.core.mt as mt
```

```python

#DATA_DIR=r'E:\GitHubz\mtpy2\examples\data'  # Windows PC

DATA_DIR=r'/g/data/ha3/fxz547/Githubz/mtpy2/examples/data'  # Linux
edi_fn = os.path.join(DATA_DIR, 'edi_files/pb40c.edi')

mt_obj = mt.MT(edi_fn)
```

## plot mt response

```python
mt_plot = mt_obj.plot_mt_response()
```

*plot phase tensors*

```python
mt_plot.plot_pt = 'y'
mt_plot.redraw_plot()
```

*plot strike angle*

```python
mt_plot.plot_strike = 'ytip'
mt_plot.redraw_plot()
```

```python
mt_plot.plot_strike = 'ytip'
mt_plot.plot_skew = 'y'
mt_plot.plot_tipper='yri'  #?? No Tipper data for station pb40

mt_plot.redraw_plot()
```
